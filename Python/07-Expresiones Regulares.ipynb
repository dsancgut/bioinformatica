# EXPRESIONES REGULARES

El módulo re contiene los métodos necesarios para analizar y buscar expresiones regulares en Python.

**import re**

**re.search(pattern,string)**

La función usa dos argumentos, el patrón y la cadena dónde hay que buscarlo


Para indicar a Python que ignore los caracteres especiales como *\t* o *\n* se puede usar *r* fuera de las comillas: **print(r”\t\n”)**

El resultado que devuelve **re.search()** es booleano, true/false, ola encuentra o no la encuentra, siempre poner import re, en parte esxtras más info.

```python
import re
```

```python
# re.search(pattern,string)
# r se usa para que Python ignore caracteres especiales
```

```python
dna = "ATCGCGAATTGAFCACGAZ"
if re.search(r"GAZ", dna):
    print("restriction enzyme site found!")
else:
    print("Not site found!")
    # la r de delante es un recurso que ignora caracteres especiales, si encuentra la exprezión nos da un mensaje y si no la encuentra nos da otro
```

```python
# Alternaciones queremos buscar más de un ptrón a la vez.
dna = "ATCGCGAATTCAC"
if re.search(r"ATT", dna) or re.search(r"CGAA", dna):
    print("restriction enzyme site found!")
```

```python
dna = "ATCGCGAATTCAC"
if re.search(r"GC(G|C)AATT", dna):
    print("restriction enzyme site found!")
    # barra vertical significa o que hay que ponero entre parebtesis y con la barr vertical
```

```python
# Grupos de caracteres
dna = "ATCGCGGCAATTCAC"
if re.search(r"GC(A|T|G|C)GC", dna):
    print("restriction enzyme site found!")
```

```python
dna = "ATCGCGGCAATTCAC"
if re.search(r"GC[ATGC]GC", dna):
    print("restriction enzyme site found!")	
    # se pueden poner las alternativas todas juntas entre corchetes es lo mismo que el de arriba
```

```python
# Negaciones, sombrerito n significa todo menos la n
dna = "ATCGCGAATTCAC"
if re.search(r"GC[^N]", dna):
    print("restriction enzyme site found!")
```

```python
# Extración de parte del patrón en una variable, esto en el trabajo, hay varios, para guadar los patrones, todo lo guardo en la variable m, la variable la podemos desglosar con el m.group
# la llave significa de repetición es decir que eso pase tres veces, \d{5} significa 5 números \s, punto puede ser cualquir cosa si queremos el punto literal sería barra punto.
# también puedo poner que se repita de tres a cinco veces {3:5}, para pillar cualquier tipo de caracter especial hay que poner la barra invertida delante.
# la llave 3 solo repite tres veces lo que hay entre corchetes, es decir pondra un GA y seguido de tres nucleótidos
dna = "ATGACGTACGTACGACTG"
# guarda los patrones en la variable m
m = re.search(r"GA[ATGC]{3}AC",dna)
m_pattern = m.group()
print(m_pattern)
```

```python
# Extrae un más de un bit del patrón
dna = "ATGACGTACGTACGACTG"
m = re.search(r"GA([ATGC]{3})AC([ATGC]{2})AC", dna)
print("entire match: " + m.group()) #me mprime todo
print("first bit: " + m.group(1)) #me imprime el primero
print("second bit: " + m.group(2)) #me impime el segundo
```

```python
# Algunos métodos
dna = "ATGACGTACGTACGACTG"
m = re.search(r"GA([ATGC]{3})AC([ATGC]{2})AC", dna)
print("input string: " + dna)
print("entire match: " + m.group())
```

```python
# Start() en que posición empieza para sacar los fragmentos de resticcion de una encima  y con end en dond acaba
print("start: " + str(m.start()))
```

```python
# End() acaba en la 12
print("end: " + str(m.end()))
```

```python
# Extraer  usando group()
dna = "ATGACGTACGTACGACTG"
m = re.search(r"GA([ATGC]{3})AC([ATGC]{2})AC", dna)
print("input string: " + dna)
print("entire match: " + m.group())
print("first bit: " + m.group(1))
print("second bit: " + m.group(2))
```

```python
# Finditer() detecta las bases erróneas, iter para ls enzimas de restriccion encontramos el patron y donde empieza
dna = "CGCTCNTAGATGCGCRATGACTGCAYTGC" 

matches = re.finditer(r"[^ATGC]", dna) 
for m in matches: 
    base = m.group() 
    pos  = m.start() 
    print(base + " found at position " + str(pos))
```

```python
dna = "CGCTCNTAGATGCGCRATGACTGCAYTGC" 

matches = re.finditer(r"[^ATGC]", dna) 

for i in matches:
    match = i.group()
    pos = i.start()
    print(f"La base {match} está en la posición {pos}")
```

```python
# Findall() util en patrones que se repiten varias veces
dna = "CTGCATTATATCGTACGAAATTATACGCGCG" 
result = re.findall(r"[AT]{6,}", dna) 
print(result)
```
