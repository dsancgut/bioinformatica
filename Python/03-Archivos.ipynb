### ARCHIVOS

Python emplea unos file objects para interactuar con los archivos de tu ordenador.

Puede manejar cualquier tipo de archivo, para algunos es posible que necesites librer√≠as adicionales.

Podemos abrir archivos con **open()**, leerlos con **read() o readlines()**, cerrarlos con **close()** y escribir en ellos con **write()**. 

Usando **open()** y w, w+, a+ nos permite sobreescribir o a√±adir informaci√≥n a la ya existente.

```python
# Vamos a abrir el archivo test.txt.
mi_archivo=open('test.txt')
```

```python
mi_archivo
```

```python
# Leemos el archivo. 
mi_archivo.read()
```

```python
# Pero, qu√© pasa si lo leemos m√°s de una vez? Que nos lleva el cursor al final
mi_archivo.read()
```

```python
# Podemos reseatear el cursor as√≠, que elcursor busque el elemnto 0, el primer elemento
mi_archivo.seek(0)
```

para volver a leer hay que poner el cursor otra vez al principio del texto

```python
mi_archivo.read()
```

```python
mi_archivo
```

```python
# Si usamos la funci√≥n readlines() nos devuelve una lista para cada l√≠nea del archivo
mi_archivo.seek(0)
mi_archivo.readlines()
```

```python
# Podemos cerrar un archivo. Siempre es buena idea cerrarlos para evitar que consuman memoria.
mi_archivo.close()
```

```python
# Podemos escribir un archivo usando open y w+ que se emplea para leer y escribir sobre el archivo.
# Hay que tener en cuenta que si se usa w o w+ se elimina el contenido anterior del archivo!!!!
mi_archivo=open('test.txt','w+')
```

```python
# Ahora escribimos en √©l, nos da 23 porque es el n√∫mero de caracters de esa nueva l√≠nea
mi_archivo.write('Esta es una linea nueva')
```

```python
mi_archivo.seek(0)
mi_archivo.read()
```

```python
mi_archivo.close()
```

```python
# Ahora queremos a√±adir informaci√≥n nueva
mi_archivo = open('test.txt','a+')
mi_archivo.write('\nEsta linea va detras de la anterior')
mi_archivo.write('\nY aqui ponemos otra')
```

```python
mi_archivo.seek(0)
mi_archivo.read()
```

```python
mi_archivo.seek(0)
print(mi_archivo.read())
```

```python
# Tambi√©n podemos iterar sobre los archivos
mi_archivo.seek(0)
for line in mi_archivo:
    print(line)
```

## EXPLICACI√ìN:
Abrir el archivo en modo 'a+'
mi_archivo = open('test.txt', 'a+')


'a+' significa:

a ‚Üí append ‚Üí a√±adir al final del archivo sin borrar lo que ya estaba.

+ ‚Üí permite leer y escribir al mismo tiempo.

Al abrir as√≠, el cursor empieza al final del archivo, no al principio. Por eso necesitas seek(0) para leer desde el principio.

2Ô∏è‚É£ Escribir en el archivo
mi_archivo.write('\nEsta linea va detras de la anterior')
mi_archivo.write('\nY aqui ponemos otra')


Esto a√±ade dos nuevas l√≠neas al archivo.

\n ‚Üí salto de l√≠nea.

üí° Nota: Cada write() devuelve el n√∫mero de caracteres escritos (como el 23 que viste antes).

3Ô∏è‚É£ Mover el cursor al inicio
mi_archivo.seek(0)


seek(0) ‚Üí mueve el cursor de lectura/escritura al principio del archivo.

Necesario si quieres leer todo el contenido despu√©s de escribir, porque tras write() el cursor queda al final.

4Ô∏è‚É£ Leer el contenido
mi_archivo.read()


Esto lee todo lo que queda desde el cursor hasta el final.

Como acabamos de hacer seek(0), lee todo el archivo.

print(mi_archivo.read())


Igual que antes, imprime todo el contenido.

5Ô∏è‚É£ Iterar sobre el archivo
mi_archivo.seek(0)
for line in mi_archivo:
    print(line)


Esto recorre l√≠nea por l√≠nea el archivo.

Es equivalente a leer todo con read(), pero √∫til si quieres procesar una l√≠nea a la vez.

seek(0) antes es necesario para volver al principio del archivo.

6Ô∏è‚É£ Resumen r√°pido del flujo

Abrir archivo en 'a+' ‚Üí leer y a√±adir al final.

write() ‚Üí a√±ade texto.

seek(0) ‚Üí volver al principio para leer.

read() ‚Üí lee todo desde la posici√≥n actual.

Iterar con for line in mi_archivo ‚Üí lee l√≠nea por l√≠nea.
