#### 5.- Devuelve la secuencia de ARN que surgir√≠a a partir de la siguiente secuencia. Recuerda que la secuencia de ARN resultante tiene que ser complementaria a la de ADN. (1 pto)

```python
def dna_to_rna_complement(dna_seq):
    # Diccionario de complementariedad ADN -> ARN
    complement = {
        "A": "U",  # A en ADN -> U en ARN
        "T": "A",  # T en ADN -> A en ARN
        "G": "C",  # G en ADN -> C en ARN
        "C": "G"   # C en ADN -> G en ARN
    }
    # Generar la secuencia complementaria de ARN
    rna_seq = "".join(complement[base] for base in dna_seq)
    return rna_seq

# Aplicar la funci√≥n a tu secuencia del gen CNTNAP2
rna = dna_to_rna_complement(secuencia)

# Mostrar resultados
print("Encabezado:", encabezado)
print("\nPrimeros 200 nucle√≥tidos de la secuencia de ARN:")
print(rna[:200])
print("\nLongitud de la secuencia de ARN:", len(rna), "nucle√≥tidos")
```

#### 6.- Realiza una funci√≥n que realice una digesti√≥n con las siguientes enzimas de restricci√≥n en nuestra secuencia. La funci√≥n tiene que devolver la siguiente informaci√≥n: (2 pto)
1. Enzimas que tienen sitios de corte en la secuencia y posici√≥n (o posiciones de corte)
2. Enzimas que no tienen sitios de corte (si las hay)
3. Resultado de la digesti√≥n: los fragmentos resultantes de nuestra secuencia
4. Indicar n√∫mero y tama√±o de cada fragmento

```python
enzR = {
    'HindIII' : 'A*AGCTT',
    'BamHI' : 'G*GATCC',
    'AluI' : 'AG*CT',
    'Sau3AI' : '*GATC',
    'EcoRI' : 'G*AATTC'
}
#aqu√≠ estamos haciendo un diccionario donde definimos cada enzima y el ast√©rico es el lugar donde corta cada enzima 
```

```python
def analizar_enzimas(seq, enzimas): #primero definmos la funci√≥n para encontrar los puntos de corte y creamos variables vac√≠as para ir guardando nuestras enzimas que cortan y que no cortan
    enz_cortan = {}
    enz_no = []

    for enz, sitio in enzimas.items():
        seq = seq.upper()
        sitio_sin = sitio.replace("*", "") #Pasamos todo a may√∫sculas para evitar confusiones y quitamos el asterisco para que se reconozca la secuencia limpia.
        pos_rel = sitio.index("*")
        posiciones = []

        start = 0
        while True:
            idx = seq.find(sitio_sin, start) #el bucle va recorriendo todas las enzimas del diccionario y sus correspondientes puntos de corte.
            if idx == -1:
                break
            posiciones.append(idx + pos_rel) #va guardando la posici√≥n donde encuentra un punto de corte en posiciones. Cuando lo encuentra, guarda la posici√≥n y sigue buscando en la siguiente, por eso se le a√±ade 1 
            start = idx + 1

        if posiciones:
            enz_cortan[enz] = posiciones
        else:
            enz_no.append(enz)

    # Resumen conciso
    print("=== Enzimas con sitios de corte ===")
    for enz, pos in enz_cortan.items():
        print(f"- {enz}: {len(pos)} sitio(s)")

    print("\n=== Enzimas sin sitios de corte ===")
    if enz_no:
        print(", ".join(enz_no))
    else:
        print("Todas las enzimas cortan la secuencia.")

    return enz_cortan, enz_no
```

```python
def generar_fragmentos(seq, enz_cortan):
    # Obtener posiciones de corte
    cortes = []
    for enz, pos_list in enz_cortan.items():
        for p in pos_list:
            cortes.append(p)
    cortes = sorted(cortes)

    # Crear fragmentos
    fragmentos = []
    prev = 0
    for pos in cortes:
        fragmentos.append(seq[prev:pos])
        prev = pos
    fragmentos.append(seq[prev:])

    tama√±os = [len(frag) for frag in fragmentos]

    print("=== Resultado de la digesti√≥n ===")
    print(f"N√∫mero total de fragmentos: {len(fragmentos)}")
    print(f"Tama√±os (pb): {tama√±os[:20]}{'...' if len(tama√±os) > 20 else ''}")  # muestra los primeros 20 tama√±os para que no se sature

    return fragmentos, tama√±os
```

#### 7.- Realiza una funci√≥n que mediante la introducci√≥n de una secuencia de ADN devuelva lo siguiente: (3pto)
1. Todos los ORFs posibles que puedan surgir de la secuencia.
2. Selecciona los 6 ORFs m√°s largos e indica si corresponden a la strand positiva o negativa.
3. Devuelve como mejor predicci√≥n, el m√°s largo de todos.
4. Calcula el porcentaje de amino√°cidos arom√°ticos de esa prote√≠na elegida y su longitud relativa a la secuencia original de ADN.

```python
from Bio.Seq import Seq

def ajustar_seq(seq):
    """Recorta la secuencia para que su longitud sea m√∫ltiplo de 3"""
    resto = len(seq) % 3
    if resto != 0:
        seq = seq[:-resto]  # elimina los √∫ltimos 1 o 2 nucle√≥tidos
    return seq

def orfs_simple(seq_dna):
    # Ajustar longitud
    seq_dna = ajustar_seq(seq_dna)
    seq_dna = Seq(seq_dna)
    orfs = []

    # Hebra positiva
    for frame in range(3):
        prot = seq_dna[frame:].translate(to_stop=False)
        for p in prot.split("*"):
            if len(p) > 0:
                orfs.append((p, "+"))

    # Hebra negativa
    seq_neg = seq_dna.reverse_complement()
    for frame in range(3):
        prot = seq_neg[frame:].translate(to_stop=False)
        for p in prot.split("*"):
            if len(p) > 0:
                orfs.append((p, "-"))

    # Ordenar por tama√±o
    orfs = sorted(orfs, key=lambda x: len(x[0]), reverse=True)

    # 6 ORFs m√°s largos
    top6 = orfs[:6]

    # Mejor ORF (m√°s largo)
    best_orf, strand = orfs[0]

    # % arom√°ticos
    aromaticos = sum(best_orf.count(aa) for aa in "FWY")
    porcentaje_arom = (aromaticos / len(best_orf)) * 100 if len(best_orf) > 0 else 0

    # Longitud relativa
    rel = (len(best_orf) * 3 / len(seq_dna)) * 100

    # Imprimir resultados
    print("6 ORFs m√°s largos:")
    for i, (orf, strand) in enumerate(top6, 1):
        print(f"{i}. Longitud {len(orf)} aa, Hebra {strand}")

    print("\nMejor ORF (m√°s largo):")
    print(f"Longitud {len(best_orf)} aa, Hebra {strand}")

    print("\nAn√°lisis de la prote√≠na predicha:")
    print(f"% arom√°ticos: {porcentaje_arom:.2f}%")
    print(f"Longitud relativa al ADN: {rel:.2f}%")

    # üîπ Devolver resultados
    return {
        "top6": [(len(o), s) for o, s in top6],
        "best_orf_seq": best_orf,
        "best_orf_len": len(best_orf),
        "strand": strand,
        "aromatic_percent": porcentaje_arom,
        "relative_length": rel
    }
```
