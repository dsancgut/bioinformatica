```python
# Ahora vamos a por un ejemplo un poco más complicado. Queremos que nos devuelva TODOS los numeros pares en un lista. 
# Si no hay números pares, nos va a devolver la lista vacía
def es_par(lista_numeros):
    # Creamos una lista vacía
    lista_pares = []
    # Parseamos cada elemento de nuestro input:
    for numero in lista_numeros:
        if numero % 2 == 0:
            # Si esto es verdad, añadimos nuestro elemento a la lista
            lista_pares.append(numero)
        else:
            pass
    return lista_pares
    
```

```python
# Vamos a crear ahora un pequeño juego. El jugador debe adivinar dónde está la letra "O"
# Empezamos creando una lista con tres elementos dónde la "O" se pone de forma aleatoria
from random import shuffle
mi_lista = [' ','O',' ']
def desordenar_lista(lista):
    shuffle(lista)
    return lista

desordenar_lista(mi_lista)
mi_lista
```

```python
# Ahora el jugador tiene que adivinar dónde está. Para ello hacemos una función
# Se está creando la variable intento.
# Se le asigna el valor '', que es una cadena vacía (string sin nada dentro).
# El while se ejecuta mientras intento NO sea '0', '1' o '2'.
# Si no inicializamos intento antes, Python daría error porque no sabría qué es intento.

Se le asigna el valor '', que es una cadena vacía (string sin nada dentro)
def adivinacion():
    intento = ''
    while intento not in ['0','1','2']:
        intento = input("Esocoje un número: 0, 1, 2: ")
    return int(intento)
```

```python
adivinacion()
```

```python
# Ahora comprobamos el intento
def comprobacion(lista,intento):
    if lista[intento] == 'O':
        print('¡Has acertado!')
    else:
        print('¡Has fallado! Suerte para la próxima')
        print(lista)
```

#### 1.- El menor de dos pares. Escribe una función que devuelva el menor de dos números, SOLAMENTE si los dos números son pares. En caso de que uno de los dos, o los dos, sean impares, devuelve el número más alto

```python
def menor_o_mayor(n1, n2):
    if n1 % 2 == 0 and n2 % 2 == 0:
        return min(n1, n2)  
    else:
        return max(n1, n2)  
```

```python
valor= menor_o_mayor(2,4)
valor
```

#### 2.- Escribe una función que tome como argumento una string de dos palabras y devuelva True si ambas empiezan por la misma letra

```python
def misma_letra(cadena):
    palabras = cadena.split()  # Divide la cadena en lista de palabras, forzamos que ambas esten en minusculas
    if len(palabras) != 2:
        return False  # Devuelve False si no hay exactamente dos palabras
    return palabras[0][0].lower() == palabras[1][0].lower()
```

```python
stringdedosletras= misma_letra("Buenas tardes")
stringdedosletras
```

.split() separa la frase en palabras usando espacios.

[0][0] obtiene la primera letra de la primera palabra.

.lower() convierte las letras a minúscula para que no importe si son mayúsculas o

#### 4.- Escribe una función que devuelva en mayúsculas la primera letra y la quinta letra de una palabra. Pista: usa capitalize()

```python
def mayus_1_y_5(palabra):
    if len(palabra) < 5:
        return "La palabra debe tener al menos 5 letras."
    
    # Convertimos toda la palabra a minúsculas primero
    palabra = palabra.lower()
    
    # Convertimos la primera y quinta letra en mayúsculas
    nueva_palabra = (
        palabra[0].upper() +          # Primera letra en mayúscula
        palabra[1:4] +                # Letras 2,3,4 sin cambios
        palabra[4].upper() +          # Quinta letra en mayúscula
        palabra[5:]                   # Resto de la palabra
    )
    
    return nueva_palabra
    # capitlize nos pone en mayusucula la primera letra de la primera palabra
```

```python
 primerayquinta= mayus_1_y_5("Papagayo")
 primerayquinta
```

#### 6.- Dada una lista de enteros, devuelve True si contiene un 3 al lado de otro 3

```python
def tiene_doble_3(lista):
    for i in range(len(lista) - 1):  # -1 para no salirnos del índice
        if lista[i] == 3 and lista[i+1] == 3:
            return True
    return False
```

```python
result= tiene_doble_3([2,3,3,4])
result
```

#### 7.- Escribe una función que busque numeros primos. 

```python
def es_primo(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):  # Solo revisa hasta la raíz cuadrada, for i in range significa que pase por el rango que le digas, sin incluir le final
        if n % i == 0:
            return False
    return True
```

```python
numero_primo= es_primo(15)
numero_primo
```

#### 8.- Juego de espías. Escribe una función que encuentre la secuencia 007 en una lista

```python
def encuentra_007(lista):
    for i in range(len(lista) - 2):  # -2 porque miramos de a 3 elementos, para forzar a quedarnis en la penultima posicion de la lista, porque si no nunca la mirariamos
        if lista[i] == 0 and lista[i+1] == 0 and lista[i+2] == 7:
            return True
    return False
```

```python
milista= encuentra_007([0,1,0,0,7])
milista
```

#### 9.- Juego de BLACKJACK. Dados tres numeros enteros entre el 1 y el 11, si su suma es menor o igual que 21, devuelve su suma. Si su suma es mayor que 21 y hay un 11, resta 10 al total de la suma. Finalmente, si con el ajuste es mayor que 21, se devuelve BUST.
##### 5,6,7 =18
##### 9,9,9 = BUST
##### 9,9,11 = 19

```python
def blackjack(a, b, c):
    total = a + b + c

    # Caso 1: Si la suma es 21 o menos, la devolvemos tal cual
    if total <= 21:
        return total

    # Caso 2: Si hay un 11 y la suma > 21, restamos 10
    if 11 in (a, b, c) and total - 10 <= 21:
        return total - 10

    # Caso 3: Si aún es mayor a 21 después del ajuste → BUST
    return "BUST"
```

```python
answer= blackjack(3,4,6)
answer
```
