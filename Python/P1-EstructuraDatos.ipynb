#### 12.- Crea un archivo mediante c√≥digo que se llame _tabla-nombres.txt_ y que est√© formado por 4 columnas y 4 filas con la siguiente formaci√≥n (1,5 ptos):
#### Primera columna: Nombre
#### Segunda columna: Apellidos (2 apellidos)
#### Tercera columna: Edad
#### Cuarta columna: Mascotas (nombre de la mascota si tienen)

#### ¬øQu√© pasa si abres el archivo desde Jupyter y has metido acentos? ¬øPuedes solucionarlo?

```python
datos = [
    ["Nombre", "Apellidos", "Edad", "Mascotas"],
    ["Ana", "Garc√≠a L√≥pez", "25", "Luna"],
    ["Juan", "P√©rez S√°nchez", "30", "Max"],
    ["Mar√≠a", "Rodr√≠guez Torres", "28", ""]
]

# Crear y escribir el archivo
with open("tabla-nombres.txt", "w", encoding="utf-8") as archivo:
    for fila in datos:
        # Unir las columnas con tabulador para simular columnas
        archivo.write("\t".join(fila) + "\n")

print("Archivo 'tabla-nombres.txt' creado con √©xito.")
```

Por qu√© "w"?

Cuando usamos open("archivo.txt", "w"), esa letra indica el modo de apertura:

"w" = write (escribir) ‚Üí Crea el archivo desde cero.
‚ö†Ô∏è Si ya exist√≠a, borra lo que ten√≠a y lo vuelve a escribir.

"r" = read (leer) ‚Üí Solo abre el archivo para leerlo.

"a" = append (a√±adir) ‚Üí Abre el archivo para agregar contenido al final sin borrar lo que ya ten√≠a.

üëâ En tu caso:

Usamos "w" para crear y escribir la tabla.

Usamos "r" despu√©s, para leer lo que ya escribimos.

üîπ ¬øPor qu√© "utf-8"?

utf-8 es una codificaci√≥n de caracteres.
Sirve para que Python sepa c√≥mo guardar y leer letras, n√∫meros y s√≠mbolos especiales.

Si escribes "Garc√≠a" o "Mu√±oz", los acentos y la √± pueden verse mal (Garc√É¬≠a) si no dices que use utf-8.

utf-8 es el est√°ndar universal: soporta acentos, √±, emojis y m√°s.

Por eso ponemos:

with open("tabla-nombres.txt", "w", encoding="utf-8")


para escribir bien los acentos.
y luego:

with open("tabla-nombres.txt", "r", encoding="utf-8")


para leerlos correctamente.

üîπ ¬øY por qu√© cambia entre "w" y "r"?

Porque una vez que ya has creado el archivo con "w", lo cierras.
Si despu√©s quieres ver lo que escribiste, lo abres otra vez pero en modo "r".

Ejemplo completo en Jupyter:

# 1Ô∏è‚É£ Crear y escribir el archivo
with open("tabla-nombres.txt", "w", encoding="utf-8") as f:
    f.write("Nombre\tApellidos\tEdad\tMascotas\n")
    f.write("Ana\tGarc√≠a L√≥pez\t25\tLuna\n")
    f.write("Luis\tMart√≠nez P√©rez\t30\t-\n")
    f.write("Mar√≠a\tFern√°ndez D√≠az\t22\tToby\n")
    f.write("Jorge\tS√°nchez Ruiz\t28\t-\n")

# 2Ô∏è‚É£ Leer el archivo
with open("tabla-nombres.txt", "r", encoding="utf-8") as f:
    print(f.read())


üëâ Resumen sencillo:

"w" = escribir (crear/reescribir).

"r" = leer.

encoding="utf-8" = para que los acentos, √± y caracteres raros se vean bien.

as f ‚Üí nombre que le damos al archivo mientras lo usamos.

\t ‚Üí tabulaci√≥n (separa columnas).

\n ‚Üí salto de l√≠nea (pasa a la siguiente fila).

#### Rattus = novergicus, rattus
#### Glaucidium = brasilianum, hoskinsii, californicum, minutissimum
#### Echinocactus = grusonii
#### Opuntia = 1


```python
dic =  {"Rattus": ["novergicus", "rattus"],
    "Glaucidium": ["brasilianum", "hoskinsii", "californicum", "minutissimum"],
    "Echinocactus": ["grusonii"],
    "Opuntia": [1]}
lista_claves = list(dic.keys())
print("Lista de claves inicial:", lista_claves)
dic["Canis"] = ["lupus", "familiaris"]
dic["Felis"] = ["catus"]
lista_claves = list(dic.keys())
set_claves = set(lista_claves)
print("Diccionario final:", dic)
print("Set de claves:", set_claves)
```
